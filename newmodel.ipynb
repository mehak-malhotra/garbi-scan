```python
import cv2
from ultralytics import YOLO

# Load YOLO Model
model = YOLO("best.pt", verbose=False)

# Load video
video_path = "input.mp4"
cap = cv2.VideoCapture(video_path)

# Global variables for ROI selection
roi_selected = False
x1, y1, x2, y2 = 0, 0, 0, 0  # Initial ROI coordinates
start_x, start_y = 0, 0

# Mouse callback function to select ROI
def select_roi(event, x, y, flags, param):
    global x1, y1, x2, y2, roi_selected, start_x, start_y

    if event == cv2.EVENT_LBUTTONDOWN:  # Left click start
        start_x, start_y = x, y

    elif event == cv2.EVENT_LBUTTONUP:  # Left click release
        x1, y1, x2, y2 = min(start_x, x), min(start_y, y), max(start_x, x), max(start_y, y)
        roi_selected = True

# Read first frame
ret, frame = cap.read()
if not ret:
    print("Error: Couldn't read video")
    cap.release()
    cv2.destroyAllWindows()
    exit()

# Show frame and allow user to select ROI
cv2.imshow("Select ROI", frame)
cv2.setMouseCallback("Select ROI", select_roi)

while True:
    cv2.imshow("Select ROI", frame)
    if roi_selected:
        cv2.destroyAllWindows()
        break
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to quit
        cap.release()
        cv2.destroyAllWindows()
        exit()

print(f"ROI Selected: ({x1}, {y1}) to ({x2}, {y2})")

# Start processing video with selected ROI
frame_count = 0
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    frame_count += 1
    print(f"Processing Frame: {frame_count}")

    # Crop to selected ROI
    roi_frame = frame[y1:y2, x1:x2]

    # Run YOLO on ROI
    results = model(roi_frame, verbose=False)

    detected = False  # Flag to check detection
    for result in results:
        for box in result.boxes:
            if box.conf[0] > 0.5:  # Confidence Threshold
                detected = True
                bx1, by1, bx2, by2 = map(int, box.xyxy[0])
                cv2.rectangle(frame, (x1 + bx1, y1 + by1), (x1 + bx2, y1 + by2), (0, 255, 0), 2)

    if detected:
        print(f"Object detected in Frame {frame_count}")

    cv2.imshow("YOLO Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Global variables
drawing = False  # True jab user area select karega
ix, iy, fx, fy = -1, -1, -1, -1

# Mouse callback function
def draw_rectangle(event, x, y, flags, param):
    global ix, iy, fx, fy, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:  # Left button press
        drawing = True
        ix, iy = x, y
        print(f"Selection started at: ({ix}, {iy})")

    elif event == cv2.EVENT_MOUSEMOVE:  # Mouse move
        if drawing:
            fx, fy = x, y

    elif event == cv2.EVENT_LBUTTONUP:  # Left button release
        drawing = False
        fx, fy = x, y
        print(f"Selection ended at: ({fx}, {fy})")
        print(f"Selected Area: Top-left ({ix}, {iy}), Bottom-right ({fx}, {fy})")

# OpenCV window setup
cap = cv2.VideoCapture(0)
cv2.namedWindow("Live Video")
cv2.setMouseCallback("Live Video", draw_rectangle)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if ix != -1 and iy != -1 and fx != -1 and fy != -1:
        cv2.rectangle(frame, (ix, iy), (fx, fy), (0, 255, 0), 2)  # Draw rectangle

    cv2.imshow("Live Video", frame)

    key = cv2.waitKey(1)
    if key == 27:  # Press ESC to exit
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# YOLO model load (Verbose off rakho taaki unwanted print na ho)
model = YOLO(r"runs\detect\train6\weights\best.pt", verbose=False)  # Nano version

# Video Load karo
video_path = "test.mp4"  # Apni video ka naam daalo
cap = cv2.VideoCapture(video_path)

# Frame-by-frame process karo
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break  # Video khatam ho gayi

    # YOLO model se detection lo (verbose=False taaki output console me na aaye)
    results = model(frame, verbose=False)

    # Sabhi detected objects par loop karo
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
            confidence = box.conf[0]  # Confidence score
            class_id = int(box.cls[0])  # Class ID

            # Box aur text draw karo
            label = f"{model.names[class_id]}: {confidence:.2f}"  # Label with confidence
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green Box
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # *Real-time display*
    cv2.imshow("Real-Time YOLO Detection", frame)

    # *Press 'q' to exit*
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Sab kuch release kar do
cap.release()
cv2.destroyAllWindows()
print("Real-Time Detection Stopped! 🛑")
```
